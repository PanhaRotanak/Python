# Music Streaming Habits Analyzer


```python
import pandas as pd

# Load datasets
history_q1 = pd.read_csv('listening_history_q1.csv')
history_q2 = pd.read_csv('listening_history_q2.csv')
track_df = pd.read_csv('track_metadata.csv')
user_df = pd.read_csv('user_demographics.csv')
```

```python
# 1. Combining Datasets
# Stitch together listening history from Q1 and Q2
history_df = pd.concat([history_q1, history_q2], ignore_index=True)

# Sort the dataframe by 'listen_id' in descending order
history_df = history_df.sort_values('listen_id', ascending=True)

# Join the 'Listening History' table with the 'Track Metadata' table
merged_df = pd.merge(history_df, track_df, left_on='trk_ref_id', right_on='track_id', how='left')

# Join with User Demographics to get 'User Age Group'
merged_df = pd.merge(merged_df, user_df, left_on='usr_ref_id', right_on='user_id', how='left')

merged_df
```

```python
# 2. Modifying DataFrames
# Rename poorly named columns
merged_df = merged_df.rename(columns={'length_in_ms': 'Milliseconds'})

# Create a new calculated column called Minutes_Played by dividing the Milliseconds column by 60,000
merged_df['Minutes_Played'] = merged_df['Milliseconds'] / 60000.0

# Use .drop() to clear redundant system ID columns
merged_df = merged_df.drop(columns=['internal_db_id', 'sys_user_hash', 'listen_id', 'trk_ref_id', 'usr_ref_id'])

merged_df.head()
```

```python
# 3. Applying Functions
# Convert timestamp to datetime
merged_df['timestamp'] = pd.to_datetime(merged_df['timestamp'])

# Write a custom function to categorize the time of day
def categorize_time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

# Apply the function via .apply()
merged_df['Time_of_Day'] = merged_df['timestamp'].dt.hour.apply(categorize_time_of_day)
merged_df.head()
```

```python
# 4. Grouping and Aggregation
# Group the data by 'Genre' and use .agg() to find the total minutes played, the average song duration, and the unique count of listeners
genre_stats = merged_df.groupby('Genre').agg(
    Total_Minutes_Played=('Minutes_Played', 'sum'),
    Average_Song_Duration=('Minutes_Played', 'mean'),
    Unique_Listeners=('user_id', 'nunique')
).reset_index()

genre_stats
```

```python
# 5. Reshaping and Pivoting
# Use pd.pivot_table() to create a heat-map-ready matrix showing "User Age Group" as rows, "Music Genre" as columns, and "Total Listens" as the values.
heatmap_matrix = pd.pivot_table(
    merged_df, 
    values='track_id', 
    index='User Age Group', 
    columns='Genre', 
    aggfunc='count',
    fill_value=0,
    margins=True,       
    margins_name='Total'     
)

heatmap_matrix
```
