## Indexing and Selection: 
#### Column/Label/Integer: Use dictionary-style selection to grab the 'Location' column. Use .iloc[] to extract the top 5 largest earthquakes based on row position.

## Indexing and Selection: 
#### Filtering: Apply multiple boolean conditions using & and | to find earthquakes that are both greater than magnitude 6.0 and occurred in a specific region (e.g., "Japan" or “Texas”).

```python
print("--- Filtered: Magnitude > 6.0 AND Region is Japan or Texas ---")
condition = (df['mag'] > 6.0) & (df['place'].str.contains('Japan|Texas', case=False, na=False))
filtered_df = df[condition]
filtered_df
```

```python
# 5. Data I/O: Exporting the filtered findings
print("--- Exporting Data ---")
export_filename = 'filtered_earthquakes.json'
filtered_df.to_json(export_filename, orient='records', indent=4)
print(f"Filtered data successfully exported to '{export_filename}' for the web team.")
```
