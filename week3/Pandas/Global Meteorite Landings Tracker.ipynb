#### The Scenario: You act as junior planetary scientists analyzing historical data of meteorite impacts and discoveries on Earth.
#### Data Structures: You separate out specific columns, like the mass_grams, to see how it behaves as an isolated 1D Series compared to the broader 2D DataFrame context.

You will use .head() to see the layout of the columns, and use .value_counts() to see which meteorite classification types (e.g., Chondrite, Iron) are the most frequently found.
## Indexing and Selection: 
#### Column/Label/Integer: Use dictionary-style selection to grab the fall_status column (whether it was seen falling or just found later). Use .iloc[] to extract the top 5 heaviest meteorites based on their sorted row position.
#### Filtering: Apply multiple boolean conditions using & and | to find meteorites that are both greater than 50,000 grams in mass and were discovered in a specific region (e.g., "Antarctica" or "Sahara").

```python
import pandas as pd
from IPython.display import display

# 1. Data I/O: Loading the raw data
print("--- Loading Data ---")
df = pd.read_csv('meteorites.csv')
print("Data loaded successfully.")
```

```python
# 2. Initial Data Inspection
print("--- Data Info ---")
df.info()

print("\n--- Data Description ---")
display(df.describe())

print("\n--- First 5 Rows ---")
display(df.head())

print("\n--- Classification Value Counts ---")
display(df['classification'].value_counts().to_frame())
```

```python
# 3. Indexing and Selection
print("--- Fall Status Column (First 5) ---")
fall_status = df['fall_status']
display(fall_status.head().to_frame())

print("\n--- Top 5 Heaviest Meteorites ---")
sorted_df = df.sort_values(by='mass_grams', ascending=False)
top_5_heaviest = sorted_df.iloc[:5]
display(top_5_heaviest[['name', 'mass_grams', 'region']])
```

```python
# 4. Filtering
print("--- Filtered: Mass > 50,000g AND Region is Antarctica or Sahara ---")
condition = (df['mass_grams'] > 50000) & (df['region'].isin(['Antarctica', 'Sahara']))
filtered_df = df[condition]
display(filtered_df)
```

```python
# 5. Data I/O: Exporting the filtered findings
print("--- Exporting Data ---")
export_filename = 'filtered_meteorites.json'
filtered_df.to_json(export_filename, orient='records', indent=4)
print(f"Filtered data successfully exported to '{export_filename}' for the museum curation team.")
```
