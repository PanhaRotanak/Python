# Restaurant Economics
#### Dataset: sns.load_dataset('tips') 
#### Goal: Find out what factors drive higher tips at a restaurant.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
sns.set_theme(style="whitegrid")
```

### Num vs. Num - Scatter: Create a sns.scatterplot() with total_bill on the X-axis and tip on the Y-axis to prove that higher bills equal higher tips.

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', color="blue")
plt.title('Scatter Plot: Total Bill vs Tip')
plt.show()
```

#### English: This scatter plot illustrates the relationship between the total bill and the tip amount. Each dot represents a single dining party. There is a clear positive correlation: as the total bill increases (moving to the right on the X-axis), the tip amount also generally increases (moving up on the Y-axis). This proves that diners typically tip a percentage of their overall bill, making the bill size a primary driver for higher tips.

#### Khmer: គំនូសតាងចំណុច (Scatter Plot) នេះបង្ហាញពីទំនាក់ទំនងរវាងវិក្កយបត្រសរុប និងចំនួនទឹកប្រាក់ធីប។ ចំណុចនីមួយៗតំណាងឱ្យអតិថិជនមួយតុ។ វាមាន ទំនាក់ទំនងវិជ្ជមាន យ៉ាងច្បាស់៖ នៅពេលដែលតម្លៃវិក្កយបត្រសរុបកើនឡើង (រំកិលទៅស្តាំលើអ័ក្ស X) ចំនួនប្រាក់ធីបក៏កើនឡើងជាទូទៅដែរ (រំកិលឡើងលើលើអ័ក្ស Y)។ វាបង្ហាញថាភ្ញៀវភាគច្រើនផ្តល់ប្រាក់ធីបជាភាគរយនៃវិក្កយបត្រសរុបរបស់ពួកគេ ដែលធ្វើឲ្យទំហំវិក្កយបត្រជាកត្តាចម្បងក្នុងការទទួលបានប្រាក់ធីបខ្ពស់។

### Num vs. Num - Line: Create a mock timeline of total_bill values using sns.lineplot() to practice connecting data points.

```python
sns.lineplot(data=tips, x=tips.index, y='total_bill', color="green")
plt.title('Line Plot: Mock Timeline of Total Bill')
plt.xlabel('Customer / Order Index')
plt.show()
```

#### English: This line plot treats the data index (order of rows in the dataset) as a mock timeline, showing how the total bill fluctuates from one customer to the next. Because the data isn't actually sorted by time, the line looks very jagged and volatile. It effectively demonstrates the range and frequency of bill amounts, showing that most bills fall between $10 and $30, with occasional sharp spikes representing very large orders over $40.

#### Khmer : គំនូសតាងខ្សែបន្ទាត់ (Line Plot) នេះចាត់ទុកលំដាប់ទិន្នន័យដូចជាពេលវេលា ដើម្បីបង្ហាញពីការឡើងចុះនៃវិក្កយបត្រសរុបពីអតិថិជនម្នាក់ទៅអតិថិជនម្នាក់ទៀត។ ដោយសារទិន្នន័យមិនត្រូវបានតម្រៀបតាមពេលវេលាពិតប្រាកដ ខ្សែបន្ទាត់នេះមើលទៅមានការប្រែប្រួលខ្លាំង។ វាបង្ហាញយ៉ាងច្បាស់ពី ទំហំ និងភាពញឹកញាប់នៃចំនួនទឹកប្រាក់ ដោយបង្ហាញថាវិក្កយបត្រភាគច្រើនស្ថិតនៅចន្លោះពី ១០ ដុល្លារ ទៅ ៣០ ដុល្លារ ដោយមានការកើនឡើងខ្លាំងម្តងម្កាលដែលតំណាងឱ្យការកម្ម៉ង់ច្រើនជាង ៤០ ដុល្លារ។

### Num vs. Cat - Box Plots: Use sns.boxplot() with day on the X-axis and total_bill on the Y-axis to find the median bill and outliers for each day of the week.

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='day', palette="Set2", legend=False)
plt.title('Box Plot: Total Bill by Day of the Week')
plt.show()
```

#### English: The box plot compares the distribution of total bills across different days of the week. The horizontal line inside each colored box represents the median bill for that day. The dots outside the "whiskers" are outliers (unusually large bills). You can see that the weekends (Saturday and Sunday) not only have a higher median bill compared to Thursday and Friday, but they also have the most significant high-spending outliers.

#### Khmer: គំនូសតាងប្រអប់ (Box Plot) ប្រៀបធៀបរបាយនៃវិក្កយបត្រសរុបតាមថ្ងៃនីមួយៗក្នុងសប្តាហ៍។ បន្ទាត់ផ្តេកនៅក្នុងប្រអប់ពណ៌នីមួយៗតំណាងឱ្យ តម្លៃមធ្យម (Median) នៃវិក្កយបត្រសម្រាប់ថ្ងៃនោះ។ ចំណុចនៅក្រៅបន្ទាត់គឺជា ទិន្នន័យខុសប្រក្រតី (Outliers) (វិក្កយបត្រដែលធំជាងធម្មតា)។ អ្នកអាចមើលឃើញថាថ្ងៃចុងសប្តាហ៍ (ថ្ងៃសៅរ៍ និងអាទិត្យ) មិនត្រឹមតែមានតម្លៃមធ្យមនៃវិក្កយបត្រខ្ពស់ជាងថ្ងៃព្រហស្បតិ៍ និងថ្ងៃសុក្រប៉ុណ្ណោះទេ ថែមទាំងមានភ្ញៀវដែលចំណាយប្រាក់ច្រើនខុសពីធម្មតាច្រើនជាងគេផងដែរ។

### Num vs. Cat - Violin Plots: Use sns.violinplot() comparing time (Lunch vs. Dinner) and tip to see where the density of generous tippers lies.

```python
sns.violinplot(data=tips, x='time', y='tip', hue='time', palette="muted", legend=False)
plt.title('Violin Plot: Tip Distribution by Time (Lunch vs. Dinner)')
plt.show()
```

#### English: The violin plot combines a box plot with a density plot to show where the majority of tip amounts are concentrated for Lunch versus Dinner. The wider the "violin" shape is at a certain point, the higher the density of tips at that value. Dinner shows a longer, thicker shape towards the top, indicating that a higher density of generous tippers arrive during dinner service, and the tip amounts are much more varied than at lunch.

#### Khmer: គំនូសតាងវីយូឡុង (Violin Plot) បញ្ចូលគ្នារវាងគំនូសតាងប្រអប់ និងគំនូសតាងដង់ស៊ីតេ ដើម្បីបង្ហាញថាតើចំនួនប្រាក់ធីបភាគច្រើនប្រមូលផ្តុំនៅត្រង់ណាសម្រាប់អាហារថ្ងៃត្រង់ ធៀបនឹងអាហារពេលល្ងាច។ កន្លែងដែលរាង "វីយូឡុង" រីកធំ មានន័យថាមានចំនួនប្រាក់ធីបនោះច្រើនជាងគេ។ អាហារពេលល្ងាចមានរាងធំ និងវែងជាងទៅផ្នែកខាងលើ ដែលបង្ហាញថាមានអតិថិជនដែលផ្តល់ប្រាក់ធីបច្រើនមកទទួលទាននៅពេលល្ងាច ហើយទំហំប្រាក់ធីបមានភាពសម្បូរបែបជាងបើធៀបនឹងអាហារថ្ងៃត្រង់។

#### Num vs. Cat - Bar Plots: Use sns.barplot() to show the average tip amount grouped by size (party size), letting Seaborn calculate the mean automatically.

```python
sns.barplot(data=tips, x='size', y='tip', hue='size', palette="pastel", legend=False)
plt.title('Bar Plot: Average Tip by Party Size')
plt.show()
```

#### English: This bar plot shows the average (mean) tip amount grouped by the size of the dining party. Seaborn automatically calculates this average behind the scenes. The black vertical lines on top of the bars are "error bars" representing statistical uncertainty. The graph shows a clear upward trend: as the party size grows from 1 to 6 people, the average tip amount steadily increases.

#### Khmer: គំនូសតាងរបារ (Bar Plot) នេះបង្ហាញពី ចំនួនប្រាក់ធីបជាមធ្យម ដោយផ្អែកតាមចំនួនមនុស្សក្នុងមួយតុ។ កម្មវិធី Seaborn គណនាតម្លៃមធ្យមនេះដោយស្វ័យប្រវត្តិ។ បន្ទាត់បញ្ឈរពណ៌ខ្មៅនៅផ្នែកខាងលើនៃរបារគឺជា "Error bars" ដែលតំណាងឱ្យគម្លាតទិន្នន័យ។ គំនូសតាងនេះបង្ហាញពីនិន្នាការកើនឡើងយ៉ាងច្បាស់៖ នៅពេលដែលចំនួនមនុស្សកើនឡើងពី ១ ទៅ ៦ នាក់ ចំនួនប្រាក់ធីបជាមធ្យមក៏កើនឡើងជាលំដាប់ផងដែរ។
