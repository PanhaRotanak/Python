# Passenger Profiling - Titanic Dataset
### Dataset: sns.load_dataset('titanic') 
### Goal: Understand who was on the Titanic by exploring single variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
titanic
```

### Categorical - Bar/Count: Use sns.countplot() to show how many passengers were in First, Second, and Third class (class column).

```python
sns.countplot(data=titanic, x='class', hue='class')
plt.title('Passenger Count by Class')
plt.show()
```

English: A countplot (or bar chart) is perfect for single categorical variables. It simply counts how many rows (passengers) belong to each category. Looking at this graph, we can clearly see that the vast majority of passengers were in Third Class, which is more than First and Second classes combined. It makes comparing categories incredibly straightforward

Khmer: ក្រាហ្វរបារ គឺស័ក្តិសមបំផុតសម្រាប់អថេរប្រភេទ (Categorical variables)។ វាគ្រាន់តែរាប់ចំនួនអ្នកដំណើរដែលស្ថិតក្នុងប្រភេទនីមួយៗប៉ុណ្ណោះ។ តាមរយៈក្រាហ្វនេះ យើងឃើញយ៉ាងច្បាស់ថា អ្នកដំណើរភាគច្រើនបំផុតគឺស្ថិតក្នុងថ្នាក់ទី៣ (Third Class) ដែលមានចំនួនច្រើនជាងថ្នាក់ទី១ និងទី២ បញ្ចូលគ្នាទៅទៀត។ វាធ្វើឱ្យការប្រៀបធៀបមានភាពងាយស្រួល

### Categorical - Pie: Use plt.pie() to show the raw proportion of male vs. female passengers (sex column), discussing why the countplot might have been easier to read.

```python
sex_counts = titanic['sex'].value_counts()
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightpink'])
plt.title('Proportion of Passengers by Sex')
plt.show()
```

English: A pie chart shows the raw proportion of categories as a whole (100%). Here, it shows that males made up roughly 65% of the passengers, while females made up about 35%. Why a countplot is often better: The human brain struggles to accurately compare angles and 2D surface areas (which pie charts use). If the slices were closer in size (e.g., 52% vs 48%), it would be very hard to tell which is bigger in a pie chart. A countplot (bar chart) uses length, which allows for instant, precise visual comparison.

Khmer: ក្រាហ្វចំណិត (Pie Chart) បង្ហាញពីសមាមាត្រនៃប្រភេទនីមួយៗធៀបនឹងចំនួនសរុប (១០០%)។ នៅទីនេះ វាបង្ហាញថាបុរសមានប្រមាណ ៦៥% នៃអ្នកដំណើរ ខណៈស្ត្រីមានប្រមាណ ៣៥%។ មូលហេតុដែល Countplot ច្រើនតែល្អជាង: ព្រោះខួរក្បាលមនុស្សពិបាកក្នុងការប្រៀបធៀបមុំ និងផ្ទៃក្រឡាឱ្យបានច្បាស់លាស់។ ប្រសិនបើចំណិតមានទំហំប្រហាក់ប្រហែលគ្នា (ឧទាហរណ៍ ៥២% និង ៤៨%) វានឹងពិបាកមើលថាមួយណាធំជាង។ ចំណែក Countplot ប្រើប្រវែង ដែលជួយឱ្យយើងមើលឃើញពីភាពខុសគ្នាបានភ្លាមៗ និងច្បាស់លាស់។

### Numerical - Histograms: Use sns.histplot() to find the age distribution (age column) of the passengers.

```python
sns.histplot(data=titanic, x='age', bins=20, color='purple')
plt.title('Age Distribution (Histogram)')
plt.show()
```

English: A histogram groups continuous numerical data (like age) into "bins" or ranges (e.g., 0-5 years, 5-10 years) and counts how many passengers fall into each bin. This graph reveals that the bulk of the passengers were young adults between 20 and 30 years old. We can also see a smaller but distinct spike on the far left, indicating a noticeable number of babies and toddlers on board.

Khmer: អ៊ីស្តូក្រាម (Histogram) ចែកទិន្នន័យលេខ (ដូចជាអាយុ) ជា "ចន្លោះ" (Bins) ហើយរាប់ចំនួនអ្នកដំណើរក្នុងចន្លោះនីមួយៗ។ ក្រាហ្វនេះបង្ហាញថា អ្នកដំណើរភាគច្រើនជាយុវជនដែលមានអាយុចន្លោះពី ២០ ទៅ ៣០ ឆ្នាំ។ យើងក៏សង្កេតឃើញមានរបារលេចឡើងបន្តិចនៅផ្នែកខាងឆ្វេងបំផុត ដែលបញ្ជាក់ថាមានទារក និងកុមារតូចៗមួយចំនួននៅលើកប៉ាល់នោះដែរ។

### Numerical - KDE: Use sns.kdeplot() to look at the density of ticket prices (fare column) to see the massive skew toward cheaper tickets.

```python
sns.kdeplot(data=titanic, x='fare', fill=True, color='orange')
plt.title('Ticket Price Density (KDE)')
plt.show()
```

English: A filled KDE plot helps us clearly see how ticket prices are distributed. The smooth curve shows the overall trend of passenger fares, while the shaded area represents the number of passengers. It also highlights the strong right-skew, meaning most people paid low fares while only a few paid very expensive prices.
Fares 0–50: The graph reaches its highest point here, showing that most passengers bought cheap tickets (mostly 3rd class).
Fares 50–100: The curve drops quickly, meaning fewer people could afford higher ticket prices.
Fares 100–500+: The long tail on the right shows a small number of wealthy passengers who paid very expensive fares (mostly 1st class).

Khmer: ក្រាហ្វ KDE បែបចាក់ពណ៌ ជួយឱ្យយើងមើលឃើញការចែកចាយតម្លៃសំបុត្របានច្បាស់។ ខ្សែកោងរលូនបង្ហាញនិន្នាការទូទៅនៃតម្លៃសំបុត្រ ខណៈផ្ទៃពណ៌បង្ហាញពីចំនួនអ្នកដំណើរ។ វាក៏បង្ហាញថាមនុស្សភាគច្រើនបង់តម្លៃថោក ហើយមានតែអ្នកមានមួយចំនួនតូចប៉ុណ្ណោះដែលបង់ថ្លៃខ្ពស់។
តម្លៃ ០–៥០: ក្រាហ្វខ្ពស់បំផុត បង្ហាញថាអ្នកដំណើរភាគច្រើនទិញសំបុត្រថោក (ភាគច្រើនថ្នាក់ទី៣)។
តម្លៃ ៥០–១០០: ខ្សែកោងធ្លាក់ចុះលឿន បង្ហាញថាមានមនុស្សតិចដែលអាចទិញសំបុត្រថ្លៃជាងនេះ។
តម្លៃ ១០០–៥០០+: កន្ទុយវែងនៅខាងស្តាំ បង្ហាញពីអ្នកដំណើរអ្នកមានមួយចំនួនតូច ដែលបង់ថ្លៃសំបុត្រខ្ពស់ (ភាគច្រើនថ្នាក់ទី១)។

### Combining Hist + KDE: Overlay the KDE curve on top of a histogram for the age column using sns.histplot(data, x='age', kde=True) to see both the bins and the smooth trend.

```python
sns.histplot(data=titanic, x='age', bins=20, kde=True, color='green')
plt.title('Age Distribution (Hist + KDE)')
plt.show()
```

English: By combining the Histogram and KDE line, we can understand the data more clearly. The histogram shows the actual number of passengers in each age group, while the KDE line shows the overall trend smoothly. Together, they help us see how passenger ages are distributed from young to old.
Ages 0–10: There is a small rise on the left side, showing that there were some children and infants on the ship.
Ages 20–35: The graph reaches its highest point here, meaning most passengers were young adults and working-age people.
Ages 40–80: After the peak, the graph slowly decreases, showing that older passengers were fewer in number.

Khmer: ការប្រើ Histogram និង KDE Line ជាមួយគ្នា ជួយឱ្យយើងយល់ទិន្នន័យបានច្បាស់។ Histogram បង្ហាញចំនួនអ្នកដំណើរតាមក្រុមអាយុ ខណៈ KDE Line បង្ហាញនិន្នាការទូទៅនៃទិន្នន័យយ៉ាងរលូន។ វាជួយឱ្យយើងមើលឃើញការចែកចាយអាយុរបស់អ្នកដំណើរពីវ័យក្មេងទៅវ័យចាស់។
អាយុ ០–១០ ឆ្នាំ: ក្រាហ្វមានការកើនឡើងតិចៗ បង្ហាញថាមានកុមារ និងទារកមួយចំនួន។
អាយុ ២០–៣៥ ឆ្នាំ: នេះជាចំណុចខ្ពស់បំផុត ដែលបង្ហាញថាភាគច្រើនជាយុវជន និងមនុស្សវ័យធ្វើការ។
អាយុ ៤០–៨០ ឆ្នាំ: ក្រាហ្វចាប់ផ្តើមថយចុះ បង្ហាញថាអ្នកដំណើរវ័យចំណាស់មានចំនួនតិចជាងក្រុមអាយុផ្សេងៗ។
